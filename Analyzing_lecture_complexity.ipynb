```python
import webvtt
import pandas as pd
import fkscore
from os import walk
import matplotlib.pyplot as plt
import seaborn as sns
import re
import datetime
%matplotlib inline
_, _, filenames = next(walk('Data/sources/'))
```

```python
filenames
```

```python
col_names =  ['start', 'end', 'text','fk_score','fk_grade','granularity', 'lecture', 'week'] 
  
# create an empty dataframe 
# with columns 
```

```python
def read_vtt(file, wk, nm):
    df_frag = pd.DataFrame(columns = col_names)

    for caption in webvtt.read('Data/sources/'+file):
    #     print(caption.start)
    #     print(caption.end)
    #     print(caption.text)
        text = caption.text.replace('\n',' ')
        f = fkscore.fkscore(text)
        new_row = {'start' : caption.start
                   , 'end' : caption.end
                   , 'text' : text
                   , 'fk_score' : f.score['readability']
                   , 'fk_grade' : f.score['grade']
                   , 'granularity' : 'fragment'
                   , 'lecture' : nm
                   , 'week' : wk
                  }
        df_frag = df_frag.append(new_row, ignore_index=True)
    return df_frag
```

```python
def score_sentence(df_seg):
    df_sentence = pd.DataFrame(columns = col_names)
    begin = 1
    for row in df_frag.index:
    #     print(row)
        if begin:
            start = df_frag.loc[row,'start']
            text = ''
            begin = 0
            text += df_frag.loc[row,'text']
        else:
            text += ' ' + df_frag.loc[row,'text']
        if text[-1] in ['.','?','!']:
            f = fkscore.fkscore(text)
            new_row = {'start' : start
                       , 'end' : df_frag.loc[row,'end']
                       , 'text' : text
                       , 'fk_score' : f.score['readability']
                       , 'fk_grade' : f.score['grade']
                       , 'granularity' : 'sentence'
                       , 'lecture' : df_frag.loc[row,'lecture']
                       , 'week' : df_frag.loc[row,'week']
                      }
            df_sentence = df_sentence.append(new_row, ignore_index=True)
            begin = 1
    return df_sentence
        
```

```python
def score_lecture(df_sentence):
    df_lecture = pd.DataFrame(columns = col_names)
    begin = 1
    for row in df_sentence.index:
    #     print(row)
        if begin:
            start = df_sentence.loc[row,'start']
            text = ''
            begin = 0
            text += df_sentence.loc[row,'text']
        else:
            text += ' ' + df_sentence.loc[row,'text']
        if row == df_sentence.index[-1]:
            f = fkscore.fkscore(text)
            new_row = {'start' : start
                       , 'end' : df_sentence.loc[row,'end']
                       , 'text' : text
                       , 'fk_score' : f.score['readability']
                       , 'fk_grade' : f.score['grade']
                       , 'granularity' : 'lecture'
                       , 'lecture' : df_sentence.loc[row,'lecture']
                       , 'week' : df_sentence.loc[row,'week']
                      }
            df_lecture = df_lecture.append(new_row, ignore_index=True)
            begin = 1
    return df_lecture
```

```python
df_full = pd.DataFrame(columns = col_names)
for filename in filenames:
    match = re.match('(?P<week>w\d+)(?:\s)(?P<name>.*)(?:.vtt)',filename)
    print(match.groupdict()['week'],match.groupdict()['name'])
    df_frag = read_vtt(filename, match.groupdict()['week'],match.groupdict()['name'])
    df_sentence = score_sentence(df_frag)
    df_lecture = score_lecture(df_sentence)
    df_full = df_full.append(df_frag, ignore_index=True).append(df_sentence, ignore_index=True).append(df_lecture, ignore_index=True)
```

```python
df_full['Start Time'] = pd.to_datetime(df_full['start'], format='%H:%M:%S.%f')
df_full['End Time'] = pd.to_datetime(df_full['end'], format='%H:%M:%S.%f')
df_full['Duration'] = (df_full['End Time']-df_full['Start Time']).dt.total_seconds()
df_full['Start Time'] = pd.to_datetime(df_full['start'], format='%H:%M:%S.%f').dt.time
df_full['End Time'] = pd.to_datetime(df_full['end'], format='%H:%M:%S.%f').dt.time
```

# 2. & 3. Graph over time of fk_score, with one lecture per graph, with plotting the average of the last 30 seconds of data

```python
def plot_lecture(lec):
#     lec = 'Other Forms of Structured Data'
    df_new = df_full.loc[(df_full['lecture'] == lec) & (df_full['granularity'] == 'fragment'),].copy()
    avg_frag = df_new['Duration'].mean()
    # print(avg_frag)
    df_new['fk_rolling'] = df_new['fk_score'].rolling(int(30/avg_frag)).mean()
    fig, line = plt.subplots(figsize = (16,9))
    line = sns.lineplot(x = "start", y = "fk_score", 
                 data = df_new,label='Fragment Score') 
    line2 = sns.lineplot(x = "start", y = "fk_rolling", 
                 data = df_new,label = 'Rolling 30s Average') 
    # plt.xticks(rotation = 45)
    line.set_xticks(df_full.loc[(df_full['lecture'] == lec) & (df_full['granularity'] == 'sentence'),'start'])
    line.set_xticklabels(df_full.loc[(df_full['lecture'] == lec) & (df_full['granularity'] == 'sentence'),'start'],rotation=45)
    line.set_title('Lecture: ' + lec, fontsize = 18)
    line.set_xlabel('Sentence Start Time', fontsize = 18)
    line.set_ylabel ('Flesch Kincaid Readability Score', fontsize = 18)
    line.legend(loc='upper right', fontsize = 11);
    new_ticks = [i.get_text() for i in line.get_xticklabels()]
    plt.xticks(new_ticks[::int(len(new_ticks)/11)])
    plt.show()
    print()
```

```python
lectures = df_full.lecture.unique()
```

```python
for lec in lectures:
    plot_lecture(lec)
```
